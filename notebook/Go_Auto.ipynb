# Data Encoding

```python
# This newdf was created in excel after we obtain all the coordinates (longitude and latitude)
# for all unique Postal Codes in the original dataset. Since are creating a dashboard in PowerBi
# a csv file was needed, and now uploaded in this block
df = pd.read_csv("CBB_Listings_LongLat.csv")
```

```python

columns = [
    "listing_type",
    "dealer_id",
    "dealer_name",
     "listing_first_date",
    "dealer_postal_code",
    "Latitude",
    "Longitude",
    "stock_type",
    "vehicle_id",
    "vin",
    "mileage",
    "price",
    "model_year",
    "make",
    "model",
    "drivetrain_from_vin",
    "fuel_type_from_vin",
]
new_df = df[columns]
new_df.info()

```

```python
columns_cluster = [
    "dealer_id",
    "listing_type",
    "Latitude",
    "Longitude",
    "stock_type",
    "vin",
    "make",
    "model",
    "mileage",
    "price",
    "model_year",
]
cluster_df = df[columns_cluster]
cluster_df.info()
```

```python
import pandas as pd

# Sample boundaries for Edmonton regions - adjust as needed for precision
def assign_region(latitude, longitude):
    # Central Region (example coordinates)
    if 53.53 <= latitude <= 53.58 and -113.52 <= longitude <= -113.45:
        return 'Central'
    # North Region
    elif latitude > 53.58:
        return 'North'
    # South Region
    elif latitude < 53.53:
        return 'South'
    # East Region
    elif longitude > -113.45:
        return 'East'
    # West Region
    elif longitude < -113.52:
        return 'West'
    else:
        return 'Unknown'  # For any other coordinates outside the specified bounds

# Apply the function to your DataFrame
cluster_df['region'] = cluster_df.apply(lambda row: assign_region(row['Latitude'], row['Longitude']), axis=1)

# Display the DataFrame with regions
print(cluster_df[['dealer_id', 'Latitude', 'Longitude', 'region']].head())
```

```python
# Step 1: Define the Composite Key
# Use the relevant columns to create a composite key. The key will help identify unique vehicles across branches.
cluster_df.loc[:, 'composite_key'] = cluster_df['vin'].astype(str) + '_' + cluster_df['model_year'].astype(str) + '_' + cluster_df['make'] + '_' + cluster_df['mileage'].astype(str)

# Step 2: Sort the DataFrame by Listing Status and Date
# If 'listing_first_date' is not present, create a placeholder date.
if 'listing_first_date' not in cluster_df.columns:
    cluster_df.loc[:, 'listing_first_date'] = pd.Timestamp('1970-01-01')
else:
    cluster_df.loc[:, 'listing_first_date'] = pd.to_datetime(cluster_df['listing_first_date'])

# Check if 'listing_Active' column exists
if 'listing_Active' in cluster_df.columns:
    # Sort to prioritize active listings over sold listings and by most recent date.
    cluster_df = cluster_df.sort_values(by=['composite_key', 'listing_Active', 'listing_first_date'], ascending=[True, False, False])
else:
    # If 'listing_Active' is missing, sort only by composite key and listing date.
    cluster_df = cluster_df.sort_values(by=['composite_key', 'listing_first_date'], ascending=[True, False])

# Step 3: Remove Duplicate Listings within a Specified Time Window
# Define a time window (e.g., X days) to identify vehicles that are relisted after a short period.
time_window = pd.Timedelta(days=30)

# Use a custom function to identify duplicates within the time window.
def filter_recent_duplicates(df):
    unique_listings = []
    last_seen = {}

    for idx, row in df.iterrows():
        key = row['composite_key']
        listing_date = row['listing_first_date']

        if key not in last_seen or (listing_date - last_seen[key]) > time_window:
            unique_listings.append(idx)
            last_seen[key] = listing_date

    return df.loc[unique_listings]

# Apply the function to filter out the duplicates.
cluster_df = filter_recent_duplicates(cluster_df)

# Step 4: Reset Index and Clean Up
# Reset index after filtering.
cluster_df.reset_index(drop=True, inplace=True)

# Drop the 'composite_key' column if no longer needed.
cluster_df.drop(columns=['composite_key'], inplace=True)
```

```python
cluster_df.info()
```

```python
# Apply one-hot encoding only if columns exist
if 'listing_type' in cluster_df.columns:
    cluster_df = pd.get_dummies(cluster_df, columns=['listing_type'], prefix="listing", dtype=int)

if 'stock_type' in cluster_df.columns:
    cluster_df = pd.get_dummies(cluster_df, columns=['stock_type'], prefix="car", dtype=int)

# Encode only the top 10 makes
top_makes = cluster_df['make'].value_counts().nlargest(10).index
cluster_df['make_filtered'] = cluster_df['make'].apply(lambda x: x if x in top_makes else 'Other')
cluster_df = pd.get_dummies(cluster_df, columns=['make_filtered'], prefix='make', dtype=int)

# Drop 'make' column as it's now encoded
cluster_df.drop(columns=['make'], inplace=True, errors='ignore')

# Encode the region column
if 'region' in cluster_df.columns:
    cluster_df = pd.get_dummies(cluster_df, columns=['region'], prefix='region', dtype=int)

# Display the DataFrame to verify encoding
cluster_df.info()
```
